## Passa ou não passa?

Variáveis preditoras

- gre: nota no teste gre
- gpa: nota no teste gpa
- rank: ranking da universidade

### Ajuste

```r
mydata = read.csv("https://stats.idre.ucla.edu/stat/data/binary.csv")
head(mydata)
```

```r
mydata$rank = factor(mydata$rank)
mylogit = glm(admit ~ gre + gpa + rank, data = mydata, family = "binomial")
```

### Predizendo

```r
predict(mylogit, newdata = data.frame(gre = 500, gpa = 3.1, rank = as.factor(1)), type = "response")
```

## Analisando a função logística

### Parâmetros

```r
summary(mylogit)
```

### Intervalo de confiança dos parâmtetros

```r
confint(mylogit)
```

### Razão de probabilidade (Odds ratio)

```r
exp(cbind(OR = coef(mylogit), confint(mylogit)))
```

Exemplo de interpretação do OR (chance).

Se a nota GPA subir em um ponto, a chance the ser adimitido na universidade (versus não ser admitido) aumenta em um fator de 2,23.
